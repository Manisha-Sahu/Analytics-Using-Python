```python
def info_of_numeric(col):
    print(f"The mean of column {col} is {dt[col].mean()}")
    print(f"The median of column {col} is {dt[col].median()}")
    print(f"The mode of column {col} is {dt[col].mode()[0]}")
    print(f"The std of column {col} is {dt[col].std()}")
```
